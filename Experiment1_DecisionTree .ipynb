```python
df.diagnosis = [1 if i == "M" else 0 for i in df.diagnosis]
```

```python
x = df.drop(["diagnosis"], axis = 1)
y = df.diagnosis.values
```

```python
x.head()
```

```python
#Normalization
x = (x - np.min(x)) / (np.max(x) - np.min(x))
x
```

```python
from sklearn.model_selection import train_test_split
#for checking testing results
from sklearn.metrics import classification_report,confusion_matrix
#for visualising the results
from sklearn.tree import plot_tree
#80-20%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

print("training split input", x_train.shape)
print("testing split input", x_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
#score(X_train,Y_train) is measuring the accuracy of the model against the training data.
#prediction
dt.score(x_test, y_test)
```

```python
# make a class prediction using testing set
y_pred_class = dt.predict(x_test)
```

```python
print("classification report -\n",classification_report(y_test, y_pred_class))
```

```python
cm=confusion_matrix(y_test,y_pred_class)
cm
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(data=cm,linewidths=1.0, square=True, annot=True, cmap='Blues')
plt.ylabel("Actual Label")
plt.xlabel("Predicted Label")
all_sample_title='Accuracy Score: {0}'.format(dt.score(x_test,y_test))
plt.title(all_sample_title, size=15)
```
